```python
import cv2
import torch
import pandas as pd
import numpy as np
import random 
from random import shuffle



def video_dataloader(dataset, batch_size=1, shuffle_flag=True, N = 20):
  number_movie = [i for i in range(len(dataset))]
  dict_caption = {i:dataset['caption'].iloc[i] for i in range(len(dataset))}

  num_batch  = int(len(number_movie)/batch_size)

  if(len(number_movie)%batch_size != 0):
    print('your batch size in not valid')
    # break

  #shuffling data
  if(shuffle_flag):
    shuffle(number_movie)

  #list of output data
  # data_output  = []
  # label_output = []
  my_output = []
  output_dict = {'batch':0, 'label':0}


  for i in range(num_batch):
    batch_frame   = []
    batch_caption = []

    for movie_num in number_movie[i*batch_size:(i+1)*batch_size]:

      # Import the video and cut it into frames.
      vid = cv2.VideoCapture(dataset['path'].iloc[movie_num])

      num_frame          = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
      selected_frame_num = random.sample(range(num_frame), N)

      #read selected frame and creat a numpy array
      frames = []
      check = True
      j = 0
      while check:
          check, arr = vid.read()
          if j in selected_frame_num:  # This line is if you want to subsample your video
                          # (i.e. keep one frame every 20)
              frames.append(arr)
          j += 1
      frames = np.array(frames) 

      batch_frame.append(frames)
      batch_caption.append(dataset['caption'].iloc[movie_num])

    # data_output.append(torch.from_numpy(np.array(batch_frame)))
    # label_output.append(torch.from_numpy(np.array(batch_caption)))
    output_dict['batch'] = torch.from_numpy(np.array(batch_frame))
    output_dict['label'] = np.array(batch_caption)
    my_output.append(output_dict.copy())
  return my_output





##=========== Here we try to test our dataloader ==========##
num_frame    = 20
dataset      = pd.read_csv('dataset.csv')

dataloader = video_dataloader(dataset, batch_size=2, shuffle_flag=True, N=num_frame)

for data_batch_num, data_batch in enumerate(dataloader):
  
  print(data_batch_num, '---', data_batch['batch'], data_batch['label'])
```
